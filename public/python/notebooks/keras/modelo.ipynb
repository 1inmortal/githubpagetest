```python
# Forzar backend JAX antes de importar Keras
import os
os.environ["KERAS_BACKEND"] = "jax"

# Importaciones compatibles con Keras 3 (backend-agnstico)
import numpy as np
import matplotlib.pyplot as plt
from keras import ops
from keras import backend as K
from keras.models import Sequential
from keras.layers import Dense

print("Backend Keras:", K.backend())
```

```python
# ejercicio: generar datos de una función no lineal con ruido
X = np.linspace(1, 100, 100).astype("float32")
X = X.reshape(-1, 1)  # (100, 1) para alimentar una capa Dense
noise = np.random.randn(100).astype("float32")
y = (X[:, 0] ** 2 + noise).astype("float32")  # (100,)
```

```python
plt.plot(X[:, 0], y)
plt.title("Datos: y = x^2 + ruido")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
model = Sequential()
model.add(Dense(128, input_shape=(1,), activation="relu"))
model.add(Dense(1))
```

```python
model.compile(loss="mse", optimizer="adam", metrics=["mae"])
```

```python
history = model.fit(X, y, epochs=500, batch_size=20, verbose=0)
```

```python
# entrenamiento ya realizado en la celda anterior (history)
# aquí podrías inspeccionar history.history si lo deseas
```

```python
y_pred = model.predict(X, verbose=0).reshape(-1)  # (100,)
```

```python
plt.plot(X[:, 0], y, color="red", label="real")
plt.plot(X[:, 0], y_pred, label="predicción")
plt.title("Ajuste con red neuronal (MSE)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
